```python
# Q1 The Simple Linear Regression (SLR) model aims to describe the relationship between a predictor variable X and an outcome variable Y. This relationship can be expressed in the equation form:Y=β0+β1X+ϵ

#   Predictor Variable (X): This is the independent variable that we assume can help predict Y.
#   Outcome Variable (Y): This is the dependent variable we want to predict or explain.
#   Intercept (β0): This is the constant term, representing the expected value of Y when X=0.
#   Slope (β1): This coefficient describes the relationship between X and Y. It represents the change in Y for a one-unit increase in X.
#   Error Term (ϵ): This accounts for the variability in Y that cannot be explained by X. It is often assumed to follow a normal distribution with a mean of 0 and constant variance.

#Given that ϵ follows a normal distribution, Y itself can also be viewed as a sample from a normal distribution with mean β0+β1X and the same variance as ϵ.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the simple linear regression model
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
error_std = 1  # Standard deviation of the error term

# Generate random predictor values (X) from a uniform distribution
np.random.seed(0)  # For reproducibility
X = np.random.uniform(0, 10, 100)

# Generate error terms (epsilon) from a normal distribution
epsilon = np.random.normal(0, error_std, len(X))

# Calculate the outcome variable (Y) using the linear model Y = beta_0 + beta_1 * X + epsilon
Y = beta_0 + beta_1 * X + epsilon

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Observed Data (Y)', color='blue', alpha=0.6)
plt.plot(X, beta_0 + beta_1 * X, color='red', label='True Regression Line (Y = beta_0 + beta_1 * X)')
plt.xlabel('Predictor (X)')
plt.ylabel('Outcome (Y)')
plt.title('Simple Linear Regression Model')
plt.legend()
plt.show()

# Checking if Y follows a normal distribution at each X
mean_y = beta_0 + beta_1 * X  # Expected mean for Y
std_y = error_std  # Standard deviation for Y

# Calculate and display the probability density function for a normal distribution at a sample X value
sample_X = 5
sample_Y_distribution = norm.pdf(Y, loc=beta_0 + beta_1 * sample_X, scale=error_std)
plt.figure(figsize=(10, 6))
plt.plot(Y, sample_Y_distribution, label=f'Normal Distribution of Y at X = {sample_X}', color='purple')
plt.xlabel('Outcome (Y)')
plt.ylabel('Density')
plt.title(f'Distribution of Y at X = {sample_X}')
plt.legend()
plt.show()
```

```python
# Q2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Parameters for the simple linear regression model
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
error_std = 1  # Standard deviation of the error term

# Generate random predictor values (X) from a uniform distribution
np.random.seed(0)  # For reproducibility
X = np.random.uniform(0, 10, 100)

# Generate error terms (epsilon) from a normal distribution
epsilon = np.random.normal(0, error_std, len(X))

# Calculate the outcome variable (Y) using the linear model Y = beta_0 + beta_1 * X + epsilon
Y = beta_0 + beta_1 * X + epsilon

# Create a DataFrame for the simulated data
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit the Simple Linear Regression model using statsmodels
model = smf.ols('Y ~ X', data=data).fit()

# Print the summary of the fitted model
print(model.summary())

# Plot the data and the fitted regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], label='Observed Data', color='blue', alpha=0.6)
plt.plot(data['X'], model.predict(data['X']), color='red', label='Fitted Regression Line')
plt.xlabel('Predictor (X)')
plt.ylabel('Outcome (Y)')
plt.title('Fitted Simple Linear Regression Model')
plt.legend()
plt.show()
```

```python
# Q3
# Sampling Variability: Each time we simulate a dataset using the theoretical model, the randomly generated error term creates slightly different values of Y for the same values of X. Consequently, when we fit a model to this data, the estimated slope and intercept will vary slightly, leading to a different fitted line each time.
# Impact of Noise on Estimation: The true regression line represents the ideal relationship without error, while the fitted line captures how noise and sampling variability impact the accuracy of our estimation. By comparing these lines, we can see how the presence of random noise affects the fitted model.
# Understanding Estimation Error: The repeated visualization of different simulated datasets helps reinforce how the estimated parameters (intercept and slope) fluctuate around the true values due to random noise in the data.
```

```python
# Q8
# H0:β1=0 where:β1 is the slope coefficient of the predictor variable waiting.
# If β1=0, it would indicate no linear relationship between waiting and duration.

import seaborn as sns
import statsmodels.formula.api as smf

# The "Classic" Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

linear_for_specification = 'duration ~ waiting'
model = smf.ols(linear_for_specification, data=old_faithful)
fitted_model = model.fit()
fitted_model.summary()
# After running this code, we would look at the p-value associated with the waiting coefficient in the summary output. If the p-value is less than the significance level (typically 0.05), we would reject the null hypothesis, suggesting that there is a statistically significant linear relationship between waiting and duration.

# If the p-value for waiting is low: This provides evidence against the null hypothesis, indicating that the waiting time significantly predicts eruption duration.
# If the p-value for waiting is high: This suggests insufficient evidence to reject the null hypothesis, implying that there may be no significant linear relationship between waiting and duration.
```

```python
# Q9
# The main goal here is to determine if there is a statistically significant relationship between duration and waiting time for subsets of the data where the waiting time is below certain thresholds (short_wait_limit values of 62, 64, and 66 minutes).
# In the full dataset, it was observed that when the waiting time exceeds around 63 minutes, the eruption duration tends to increase notably. By restricting the dataset to short wait times (less than 62, 64, and 66 minutes), we aim to see if this relationship still holds or if the linear association between duration and waiting diminishes for shorter waiting times.

# p-value > 0.1: No evidence against the null hypothesis, indicating no significant relationship between waiting and duration within this short wait time range.
# 0.05 < p-value ≤ 0.1: Weak evidence against the null hypothesis, suggesting a slight indication of a relationship.
# 0.01 < p-value ≤ 0.05: Moderate evidence against the null hypothesis, implying a possible significant relationship.
# p-value ≤ 0.01: Strong evidence against the null hypothesis, indicating a significant linear relationship between waiting and duration.
```

```python
# Q10
# a
import seaborn as sns
import statsmodels.formula.api as smf
import numpy as np
import scipy.stats as stats  # Import scipy.stats to access the norm function
import matplotlib.pyplot as plt

# Load the Old Faithful dataset and filter for long wait times
old_faithful = sns.load_dataset('geyser')
long_wait = old_faithful['waiting'] >= 63
data_long_wait = old_faithful[long_wait]

# Bootstrapping
n_bootstrap = 1000  # Number of bootstrap samples
bootstrap_slope_coefficients = []

for _ in range(n_bootstrap):
    # Generate a bootstrapped sample
    bootstrap_sample = data_long_wait.sample(n=len(data_long_wait), replace=True)
    # Fit the model and collect the slope
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slope_coefficients.append(model.params['waiting'])

# Visualize the bootstrapped sampling distribution
plt.hist(bootstrap_slope_coefficients, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.title('Bootstrapped Sampling Distribution of Slope Coefficients')
plt.show()

# b
# Null hypothesis simulation parameters
beta_0_null = 1.65
beta_1_null = 0
std_dev = 0.37  # Standard deviation for noise

simulated_slope_coefficients = []

for _ in range(n_bootstrap):
    # Create a copy of the long wait dataset
    old_faithful_simulation = data_long_wait.copy()
    # Simulate duration under the null hypothesis
    old_faithful_simulation['duration'] = (
        beta_0_null + beta_1_null * old_faithful_simulation['waiting'] +
        stats.norm(loc=0, scale=std_dev).rvs(size=len(data_long_wait))
    )
    # Fit the model and collect the slope
    model_null = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    simulated_slope_coefficients.append(model_null.params['waiting'])

# Visualize the null hypothesis sampling distribution
plt.hist(simulated_slope_coefficients, bins=30, edgecolor='k', alpha=0.7, color='red')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.title('Sampling Distribution of Slope Coefficients Under Null Hypothesis')
plt.show()

# c
# 95% Confidence Interval for bootstrap slopes
bootstrap_conf_interval = np.quantile(bootstrap_slope_coefficients, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval: [{bootstrap_conf_interval[0]:.4f}, {bootstrap_conf_interval[1]:.4f}]")

# Check if 0 is within the confidence interval
contains_zero = bootstrap_conf_interval[0] <= 0 <= bootstrap_conf_interval[1]
print("Contains 0 in 95% CI:", contains_zero)

# Calculate the observed slope from the original long wait dataset
observed_slope = smf.ols('duration ~ waiting', data=data_long_wait).fit().params['waiting']

# Simulated p-value: proportion of null hypothesis slopes as extreme as the observed slope
simulated_p_value = np.mean([abs(slope) >= abs(observed_slope) for slope in simulated_slope_coefficients])
print(f"Simulated p-value: {simulated_p_value:.4f}")

# Compare with original model p-value
original_model_p_value = smf.ols('duration ~ waiting', data=data_long_wait).fit().pvalues['waiting']
print(f"Original model p-value: {original_model_p_value:.4f}")
```

```python
# Q13
# Permutation Testing:
# Randomizes group labels to simulate a scenario where the null hypothesis (no difference) is true.
# Tests if the observed difference is extreme under the null hypothesis.
# Provides a p-value indicating the likelihood of the observed result if there’s no true group difference.
# Bootstrapping:
# Resamples the data with replacement to estimate the sampling distribution of the mean difference.
# Provides a confidence interval for the mean difference, showing a range of plausible values for the actual difference.
# Does not test a hypothesis directly but rather quantifies uncertainty in the estimate.

# Permutation Test: Best for testing the existence of a difference without making assumptions about the data's distribution. It provides a p-value but no confidence interval or specific estimate of the difference.
# Bootstrap Confidence Interval: Best for estimating the plausible range of the difference without assumptions. It provides a confidence interval but no p-value for hypothesis testing.
# Indicator Variable Model: Best for estimating the difference and testing significance when assumptions of linearity and normality are reasonable. It provides both an estimate and a p-value, making it powerful but sensitive to assumption violations.
```
