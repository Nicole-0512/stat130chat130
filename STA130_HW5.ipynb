```python
# Null Hypothesis (H₀): This is the default assumption or claim that there is no effect or no difference. 
# It represents the status quo or what we are trying to disprove. 
# For example, if you are comparing the effectiveness of two presidents, your null hypothesis might be that there is no difference in their performance. Technically, we "fail to reject" the null hypothesis if the data does not provide strong enough evidence against it; we do not "accept" it.
# Alternative Hypothesis (H₁ or Hₐ): This is what you are trying to prove or find evidence for. 
# It represents a claim that contradicts the null hypothesis. In the presidential example, the alternative hypothesis might be that there is a significant difference in the performance of the two presidents.
```

```python
# p-value measures the probability of obtaining a test statistic as extreme as the one observed, or more extreme, if the null hypothesis were true.
```

```python
#Q1
# A testable idea (or hypothesis) is one that can be formulated in such a way that it makes a specific, measurable, and falsifiable prediction. 
# An idea that cannot be tested is one for which no concrete or measurable evidence can be gathered, or if the idea itself is too vague, subjective, or speculative to be analyzed objectively. These ideas often fall into the realm of opinion, belief, or future speculation without any supporting data.
```

```python
#Q2 
# The null hypothesis is the default assumption or status quo that there is no effect or no difference. It assumes that any observed effect in the data is due to random chance or sampling error.
# The alternative hypothesis is the opposite claim to the null hypothesis. It suggests that there is an effect or a difference. This is what we aim to provide evidence for through our test.
# The purpose of setting up a null hypothesis is to create a neutral statement that we can challenge with data, helping us assess whether there is enough evidence to support the alternative hypothesis.
# A good null hypothesis is testable, reflects the status quo, is specific, and provides a clear basis for statistical evaluation.
```

```python
#Q3
# Null Hypothesis H0:The null hypothesis is a formal statement that assumes the population mean μis equal to a specific value μ0. In other words, under H0, we assume there is no difference between the population mean and the hypothesized value μ0.
# Alternative Hypothesis Ha:The alternative hypothesis, Ha, is the complement of the null hypothesis. It states that H0 is false, meaning the population mean μis not equal to μ0.
# Observed sample values xi:These are the individual data points observed in the sample, which we use to calculate the sample mean x(hat). These values represent a small portion of the population and are used to make inferences about μ.
# Sample average x(hat):This is the average of the observed sample values, and it serves as an estimate of the population mean μ. While x(hat)is calculated from the sample, we use it to make conclusions about the overall population.
# Actual value of μ:This is the true but often unknown mean of the entire population. Our goal in hypothesis testing is to draw conclusions about μ based on sample data.
# Hypothesized value μ0:This is the specific value of the population mean assumed under the null hypothesis. In hypothesis testing, we compare the sample mean x(hat)to μ0to determine if there's enough evidence to reject H0.
```

```python
#Q6
# In this case, 80 out of 124 couples, or 64.5%, tilted their heads to the right. We will compute the p-value for this scenario using a binomial test based on the null hypothesis of a 50% tilt probability.
# hypotheses: Null Hypothesis H0: The probability of tilting the head to the right is 50%, i.e., p=0.5.
# Alternative Hypothesis HA: The probability of tilting the head to the right is not 50%, i.e.,p=/0.5.

from scipy import stats

# Parameters
n = 124  # Total number of couples
k = 80   # Number of couples tilting heads to the right
p_null = 0.5  # Null hypothesis probability of tilting right

# Perform a binomial test
p_value = stats.binom_test(k, n, p_null, alternative='two-sided')
p_value

# The computed p-value for the binomial test is approximately 0.0016.

# interpretation:
# Based on the p-value table provided:0.0016falls in the range 0.01>=p>0.001, which corresponds to strong evidence against the null hypothesis.
# Thus, there is strong evidence to suggest that humans do not tilt their heads left or right with equal probability when kissing, based on this data.
```

```python
#Q9
# Sample Size: The original Fisher-Bristol experiment had a small sample size of 8 cups, with Bristol needing to identify all correctly. In contrast, our STA130 experiment involves 80 students, each making an independent choice.
# Population: In the original experiment, the focus was on Dr. Bristol’s specific ability, which was a more personalized test of her skill. In our case, we are examining a general ability among a larger group of students, which might reflect a more abstract notion of ability or simply guessing.
# Hypothesis: Both experiments aim to test whether the observed results could be explained by random guessing or whether there’s evidence that participants are identifying the correct pouring order better than chance.

# Null Hypothesis H0: The students are randomly guessing, meaning the probability of correctly identifying the order is 50%. Formally, this is:H0:p=0.5 where p is the proportion of students correctly identifying the order.
# Alternative Hypothesis Ha: The students are not randomly guessing, suggesting that the proportion of correct identifications is greater than 50%. Formally:Ha:p>0.5

#Quantitative Analysis
#To analyze this experiment, we will perform a hypothesis test for proportions. Specifically, we will compute a z-test for the observed proportion of correct identifications compared to the expected proportion under the null hypothesis.
# Sample Size (n): 80
# Observed Proportion (p(hat)): 49/80 = 0.6125
# Null Hypothesis Proportion (p0): 0.5
# The test statistic for a proportion z-test is computed as:
# z=(p(hat)-p0/√(p0(1-p0)/n))
# where:p^is the sample proportion.
#       p0is the hypothesized proportion (0.5 under the null hypothesis).
#       n is the sample size.

# z-test
import numpy as np
from scipy import stats

# Set seed for reproducibility
np.random.seed(42)

# Parameters
n = 80  # sample size
p_hat = 49 / 80  # observed proportion of correct identifications
p_0 = 0.5  # null hypothesis proportion

# Compute the test statistic
z_stat = (p_hat - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)

# Compute the one-sided p-value
p_value = 1 - stats.norm.cdf(z_stat)

z_stat, p_value

# The purpose of the z-test for proportions is to evaluate whether the observed proportion of correct identifications (61.25%) is significantly different from the null hypothesis value of 50%, which represents random guessing. By calculating the z-statistic, we measure how far the observed proportion deviates from 50%, standardized by the expected variability under the null hypothesis.
# The z-statistic allows us to determine the likelihood of observing a proportion as extreme as 0.6125 if the null hypothesis were true. This likelihood is represented by the p-value, which tells us whether the observed result can reasonably be attributed to chance or if it provides strong evidence to reject the null hypothesis in favor of the alternative hypothesis.
# In this case, the one-tailed test is appropriate because we are only interested in whether the students are better than random guessing (i.e., if their proportion of correct answers is greater than 50%).

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution
x = np.linspace(0.4, 0.7, 100)
y = stats.norm.pdf(x, loc=0.5, scale=np.sqrt(0.5 * 0.5 / n))
plt.plot(x, y, label='Null Hypothesis Distribution')
plt.axvline(p_hat, color='red', linestyle='--', label='Observed Proportion')
plt.title("Sampling Distribution Under Null Hypothesis")
plt.xlabel('Proportion of Correct Identifications')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

# Findings and Discussion
# Z-Statistic: The z-statistic measures how many standard deviations away our observed proportion (61.25%) is from the hypothesized value (50%).
# P-Value: The p-value tells us the probability of observing a result as extreme as our observed proportion if the null hypothesis were true (i.e., if the students were simply guessing).
# The p-value will determine if the observed success rate (61.25%) is significantly greater than 50%, suggesting that students are better than random guessing at identifying the pouring order.

# Conclusion Regarding the Null Hypothesis
# Based on the calculated z-statistic and p-value, we assess whether to reject the null hypothesis. 
# If the p-value is smaller than the chosen significance level (e.g., 0.05), we will reject the null hypothesis, concluding that the students are better than random guessing. 
# If the p-value is larger, we will fail to reject the null hypothesis, indicating insufficient evidence to claim that the students have a real ability to distinguish the order of pouring.
```
