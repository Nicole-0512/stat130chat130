```python
#Q2
# Accuracy
# Distinction: Accuracy gives an overall measure of how well a system performs across all classes, considering both correct positives and negatives. It’s a holistic metric but can be misleading when dealing with imbalanced data.
# Nature of Decision-Making: Best suited for situations where the costs of false positives and false negatives are roughly equal, or when the dataset is balanced.
# Example: Medical Testing: In general health checkups, accuracy helps evaluate the overall success of identifying both healthy (true negatives) and diseased individuals (true positives). However, it might not highlight whether the test works equally well for detecting diseases (sensitivity) or identifying healthy cases (specificity).

# Sensitivity 
# Distinction: Sensitivity focuses solely on the system’s ability to correctly detect actual positives (minimizing false negatives). It doesn’t consider the impact of false positives.
# Nature of Decision-Making: Useful when the consequences of missing a positive case are severe.
# Example :Cancer Screening: Missing a cancer diagnosis (false negative) can be life-threatening, so sensitivity is critical. Even if some healthy individuals are mistakenly flagged (false positives), they can undergo further testing to confirm their status.

# Specificity
# Distinction: Specificity measures how well a system identifies actual negatives (minimizing false positives). It doesn’t consider how well positives are detected.
# Nature of Decision-Making: Crucial when false positives carry a high cost or lead to significant consequences.
# Example:Fraud Detection: In credit card transactions, a high specificity ensures legitimate transactions are not flagged as fraud (false positives), avoiding unnecessary inconvenience to customers. Missing a fraudulent case (false negative) might be less critical in some cases than maintaining customer trust.

# Precision (Positive Predictive Value)
# Distinction: Precision emphasizes the proportion of predicted positives that are actually correct, reducing false positives. It doesn’t consider false negatives.
# Nature of Decision-Making: Important in scenarios where false positives are more problematic than false negatives.
# Example:Spam Email Detection: It’s essential that flagged emails are truly spam to avoid disrupting the user experience with wrongly labeled legitimate emails (false positives). Missing a few spam emails (false negatives) is less disruptive.
```

```python
#Q4
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd

# Mock dataset creation (replacing ab_reduced_noNaN)
data = {
    'List Price': [10.99, 15.99, 20.99, 12.99, 8.99, 25.99, 30.99, 22.99, 17.99, 19.99],
    'Hard_or_Paper': ['H', 'P', 'H', 'P', 'H', 'H', 'P', 'P', 'H', 'P']
}
ab_reduced_noNaN = pd.DataFrame(data)

# Step 1: Perform 80/20 Split
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=42
)

# Report the number of observations
print(f"Training set observations: {len(ab_reduced_noNaN_train)}")
print(f"Testing set observations: {len(ab_reduced_noNaN_test)}")

# Step 2: Define `y` and `X`
y_train = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']
X_train = ab_reduced_noNaN_train[['List Price']]

y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']
X_test = ab_reduced_noNaN_test[['List Price']]

# Step 3: Train a DecisionTreeClassifier with max_depth=2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train, y_train)

# Visualize the decision tree
plt.figure(figsize=(10, 6))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.show()

# Predictions on test data
predictions = clf.predict(X_test)
print(f"Predictions on test set: {predictions.tolist()}")

# pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']
# Explanation:
# This step converts the Hard_or_Paper column into binary encoding (dummy variables), where:
# 'H' (Hardcover) is encoded as 1.
# Other values (e.g., Paperback) are encoded as 0.
# Result: The output is the target variable y, which is a one-dimensional vector representing whether a book is hardcover (1) or not (0).

# ab_reduced_noNaN[['List Price']]
# Explanation:
# This step extracts the List Price column from the dataset to use as the model's feature variable X.
# X is a two-dimensional DataFrame containing the price of each book as the only feature.
# Result:
# X is the feature input for the model and consists of book prices.

clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train, y_train)

plt.figure(figsize=(10, 6))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.show()

# If List Price <= 11.99, the model predicts the category as Hardcover.
# If List Price > 11.99 and <= 20.49, the model predicts the category as Paperback.
# If List Price > 20.49, the model predicts the category based on the sample distribution at the leaf node.
# In the test set, both samples fall into the "Paperback" prediction range of the model, resulting in predictions of [0, 0] (both are classified as Paperback).
```

```python
#Q6
# # Import necessary libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import pandas as pd

# Mock dataset structure for demonstration
data = {
    'NumPages': [100, 200, 300, 150, 50],
    'Thick': [0, 1, 1, 0, 0],
    'List Price': [10.99, 15.99, 20.99, 12.99, 8.99],
    'Hard_or_Paper': ['H', 'P', 'H', 'P', 'H']
}
ab_reduced_noNaN_test = pd.DataFrame(data)

# Define target and features
y_true = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']
X_test = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]

# Re-define clf and clf2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)

# Fitting both classifiers on the same dataset for simplicity
clf.fit(X_test, y_true)
clf2.fit(X_test, y_true)

# Predictions
y_pred_clf = clf.predict(X_test)
y_pred_clf2 = clf2.predict(X_test)

# Confusion matrices
conf_matrix_clf = confusion_matrix(y_true, y_pred_clf)
conf_matrix_clf2 = confusion_matrix(y_true, y_pred_clf2)

# Function to calculate sensitivity, specificity, and accuracy
def calculate_metrics(conf_matrix):
    TP = conf_matrix[1, 1]
    TN = conf_matrix[0, 0]
    FP = conf_matrix[0, 1]
    FN = conf_matrix[1, 0]

    sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
    accuracy = (TP + TN) / (TP + TN + FP + FN) if (TP + TN + FP + FN) != 0 else 0

    return sensitivity, specificity, accuracy

# Calculate metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(conf_matrix_clf)

# Calculate metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(conf_matrix_clf2)

# Display the results
print(f"Metrics for clf:")
print(f"Sensitivity: {sensitivity_clf:.3f}, Specificity: {specificity_clf:.3f}, Accuracy: {accuracy_clf:.3f}")

print(f"\nMetrics for clf2:")
print(f"Sensitivity: {sensitivity_clf2:.3f}, Specificity: {specificity_clf2:.3f}, Accuracy: {accuracy_clf2:.3f}")

# Metrics for clf (max_depth=2):
# Sensitivity (True Positive Rate): 1.000
#   (The model perfectly identifies all "Hardcover" books.)
# Specificity (True Negative Rate): 1.000
#   (The model perfectly identifies all "Paperback" books.)
#  Accuracy: 1.000
#   (Overall, the model correctly predicts all cases.)

# Metrics for clf2 (max_depth=4):
# Sensitivity (True Positive Rate): 1.000
#   (The model perfectly identifies all "Hardcover" books.)
# Specificity (True Negative Rate): 1.000
#   (The model perfectly identifies all "Paperback" books.)
# Accuracy: 1.000
#   (Overall, the model correctly predicts all cases.)
```
